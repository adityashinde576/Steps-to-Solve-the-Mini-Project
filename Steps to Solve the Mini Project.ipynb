```python
import math
def sigmoid(z):
    return 1 / (1 + math.exp(-z))
def relu(z):
    return max(0, z)
def softmax(z_list):
    exp_vals = [math.exp(z) for z in z_list]
    total = sum(exp_vals)
    return [e / total for e in exp_vals]
def mse_loss(y_true, y_pred):
    return 0.5 * (y_true - y_pred) ** 2
def bce_loss(y_true, y_pred):
    return -(y_true * math.log(y_pred) + (1 - y_true) * math.log(1 - y_pred))
def cce_loss(y_true_list, y_pred_list):
    loss = 0
    for y_t, y_p in zip(y_true_list, y_pred_list):
        loss += -y_t * math.log(y_p)
    return loss
W1 = [0.8, -0.4]
B1 = [0.2, 0.1]
W2 = [1.0, -1.2]
B2 = 0.3
def forward_pass(x, y_true):
    print("Input:", x)
    z1_1 = W1[0] * x + B1[0]
    z1_2 = W1[1] * x + B1[1]
    print("z1:", z1_1, z1_2)
    a1_1 = sigmoid(z1_1)
    a1_2 = sigmoid(z1_2)
    print("a1:", a1_1, a1_2)
    z2 = a1_1 * W2[0] + a1_2 * W2[1] + B2
    print("z2:", z2)
    y_hat = sigmoid(z2)
    print("y_hat:", y_hat)
    loss = mse_loss(y_true, y_hat)
    print("Loss:", loss)
    print("---------------------------------")
    return y_hat, loss
data = [
    (0, 0),
    (1, 0),
    (0, 0),
    (1, 1)
]

for x, y in data:
    forward_pass(x, y)
```
